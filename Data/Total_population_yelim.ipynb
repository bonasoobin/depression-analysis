```python
# 초기화 
new_pop['도'] = None
new_pop['시'] = None
new_pop['군/구'] = None

# 행정명에서 도, 시, 군/구로 파생변수 생성
def assign_region(row):
    parts = row['행정명'].split()
    for part in parts:
        if part.endswith('도'):
            row['도'] = part
        elif part.endswith('시'):
            row['시'] = part
        elif part.endswith('군') or part.endswith('구'):
            row['군/구'] = part
    return row

new_pop = new_pop.apply(assign_region, axis=1)
new_pop.info()
```

### **시도, 시군구 컬럼 추가**

```python
# 특별시 
new_si = ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시']

# '시도' 컬럼
new_pop['시도'] = new_pop.apply(lambda row: row['시'] if row['시'] in new_si else row['도'], axis=1)
```

```python
new_pop['시도'].unique()
```

```python
# '시군구' 컬럼을 생성
def sigungu(row):
    if row in ['충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도']:
        return row[0] + row[2]
    elif row in ['서울특별시', '경기도', '강원도', '제주특별자치도', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시',
       '세종특별자치시']:
        return row[:2]
    else:
        return row  # 나머지 경우에는 원본 값

# '시군구' 컬럼 생성
new_pop['시군구'] = new_pop['시도'].apply(sigungu)
```
