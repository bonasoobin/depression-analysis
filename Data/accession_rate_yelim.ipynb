```python
import pandas as pd
import numpy as np
```

```python
accession_rate = pd.read_csv('raw_data/시군구_경제활동인구_총괄_20231219121226.csv',  encoding = 'CP949')
print(accession_rate.head())
```

```python
accession_rate = accession_rate[1:]
accession_rate.head()
```

```python
# 파생 변수 생성 함수 정의
def create(df, start_year, end_year):
    for year in range(start_year, end_year + 1):
        col1 = f"{year}.1/2"
        col2 = f"{year}.2/2"

        # 두 열 모두 존재하는 경우 평균 계산
        if col1 in df.columns and col2 in df.columns:
            df[col1] = df[col1].astype(float)
            df[col2] = df[col2].astype(float)
            df[str(year)] = df[[col1, col2]].mean(axis=1)
        # 한 열 또는 두 열이 누락된 경우 해당 연도의 열을 0으로 설정
        else:
            df[str(year)] = 0

    return df

# 함수 적용
new_accession_rate= create(accession_rate, 2018, 2022)

# 업데이트된 데이터프레임 확인
new_accession_rate.head()
```

```python
new_accession_rate = new_accession_rate.drop(new_accession_rate.columns[1:11], axis=1)
new_accession_rate
```

```python
# CSV 파일로 저장
new_accession_rate.to_csv('accession_rate.csv', index=False,  encoding = 'CP949')
```
