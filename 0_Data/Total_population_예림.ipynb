#### **행정구역을 기준으로 2018 ~ 2022 데이터 join**

```python
pop_list = [pop_2018, pop_2019, pop_2020, pop_2021, pop_2022]
new_pop = reduce(lambda x, y : pd.merge(x, y, on='행정구역'), pop_list)
new_pop
```

```python
new_pop.info()
```

```python
# '행정구역' 컬럼을 제외한 모든 컬럼 쉼표 제거 후 숫자형으로 변환
for col in new_pop.columns:
    if col != '행정구역':
        new_pop[col] = new_pop[col].str.replace(',', '').astype(int)

new_pop.info()
```

```python
# '행정구역' 컬럼 분리(정규표현식)
new_pop[['행정명', '행정코드']] = new_pop['행정구역'].str.extract(r'^(.*?)\s*\((\d+)\)$')
new_pop[['행정명', '행정코드']]
```

```python
new_pop.info()
```

```python
new_pop = new_pop[['행정명', '행정코드', '2018년_총인구수', '2019년_총인구수', '2020년_총인구수', '2021년_총인구수', '2022년_총인구수']]
new_pop.info()
```

```python
new_pop.head()
```

```python
# Initialize new columns
new_pop['도'] = None
new_pop['시'] = None
new_pop['군/구'] = None

# Function to assign values to new columns based on the last character
def assign_region(row):
    parts = row['행정명'].split()
    for part in parts:
        if part.endswith('도'):
            row['도'] = part
        elif part.endswith('시'):
            row['시'] = part
        elif part.endswith('군') or part.endswith('구'):
            row['군/구'] = part
    return row

new_pop = new_pop.apply(assign_region, axis=1)
new_pop.info()
```

```python
new_pop
```

```python
new_pop = new_pop[['도', '시', '군/구', '행정코드','2018년_총인구수','2019년_총인구수', '2020년_총인구수', '2021년_총인구수', '2022년_총인구수']]
new_pop.info()
```

```python
# 세종특별자치시는 두 컬럼이 똑같은 행정명이므로 중복되는 첫번째 컬럼 삭제 
new_pop = new_pop[new_pop['행정코드'] != '3600000000']
new_pop.info()
```

```python
# 도단위
new_pop['도'].unique()
```

```python
do = ['경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']
new_pop = new_pop[~((new_pop['도'].isin(do)) & (new_pop['시'].isnull()) & (new_pop['군/구'].isnull()))]
new_pop.info()
```

```python
new_pop
```

```python
si = ['서울특별시','부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
       '울산광역시','수원시', '성남시', '안양시', '안산시', '고양시', '용인시', '청주시', 
      '천안시', '전주시', '포항시', '창원시']
```

```python
len(si)
```

```python
# 'si' 리스트와 정확히 일치하는 값 &'군/구' 컬럼은 null값 삭제
new_pop = new_pop[~((new_pop['시'].isin(si)) &  (new_pop['군/구'].isnull()))]
new_pop.info()
```

### **마지막 행에 총인구수 합이 일치한지 확인하기 위한 총 합계 데이터 추가**

```python
# 5개 컬럼의 합계를 계산
sum_row = new_pop[["2018년_총인구수", "2019년_총인구수", "2020년_총인구수", "2021년_총인구수", "2022년_총인구수"]].sum()

# 계산된 합계를 데이터프레임의 마지막 행으로 추가
new_pop = new_pop.append(sum_row, ignore_index=True)
new_pop.tail(1)
```
