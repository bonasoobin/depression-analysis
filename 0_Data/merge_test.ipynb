```python
import pandas as pd
import numpy as np
```

```python
yelim = pd.read_csv('merge_tot_data.csv', encoding = 'CP949')
soobin = pd.read_csv('result_0123491011.csv', encoding='utf-8-sig')
ranee = pd.read_csv('연령,스트레스인지율병합.csv', encoding = 'CP949')
```

```python
soobin = soobin.sort_values(by=['년도','행정번호'], ascending=True).reset_index(drop=True)
soobin
```

```python
yelim = yelim.sort_values(by=['년도','행정번호'], ascending=True).reset_index(drop=True)
yelim
```

```python
ranee = ranee.sort_values(by=['년도','행정번호'], ascending=True).reset_index(drop=True)
ranee
```

```python
ranee = ranee.iloc[:,[4,5]]
yelim = yelim.iloc[:,[4,5]]
```

```python
final = pd.concat([soobin, ranee, yelim], axis=1)
final
```

```python
final.isnull().sum()
```

```python
final['코로나확진자수'].unique()
```

```python
final[final['코로나확진자수'] == ' - ']
```

```python
final['코로나확진자수'] = final['코로나확진자수'].str.replace('-', '0')
final['코로나확진자수'] = final['코로나확진자수'].str.replace(',', '').astype('int64')
final.info()
```

```python
order = ['시도', '시군구', '행정번호', '년도', '우울증환자수', '평균연령', '총인구수', 
         '스트레스인지율', '고용률', '코로나확진자수', '1인당 정신건강예산(원)', '공공시설개수']
final = final[order]
final
```

```python
final.to_csv('final.csv', encoding = 'CP949', index = False)
```
