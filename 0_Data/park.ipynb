### 공원데이터

```python
import pandas as pd
data = pd.read_csv(r'C:\Users\soobin\coding_excuse\전국도시공원정보표준데이터.csv')
data.info()
```

```python
# 공원명, 소재지도로명주소, 소재지지번주소, 지정고시일 컬럼만 선정
data1 = data.copy()
data1 = data1[data1['공원구분'] != '묘지공원']
data1 = data1.iloc[:, [1,3,4,13]]
data1
```

```python
# 소재지지번주소를 ' '으로 구분해서 각각 시도, 시군구, 동에 저장
data2 = data1.copy()
data2['시도'] = data2['소재지지번주소'].str.split(' ').str[0]
data2['시군구'] = data2['소재지지번주소'].str.split(' ').str[1]
data2['동'] = data2['소재지지번주소'].str.split(' ').str[2]

# 경기도 용인시 수지구 같은 경우는 '동' 컬럼에 수지구가 저장되므로, '동' 컬럼이 구로 끝난다면 시군구에 이어붙이기
for index, row in data2.iterrows():
    if isinstance(row['동'], str) and row['동'].endswith('구'):
        data2.at[index, '시군구'] = row['시군구'] + row['동']
        data2.at[index, '동'] = ''
```
